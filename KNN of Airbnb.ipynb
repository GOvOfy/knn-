```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

features=['accommodates','bedrooms','bathrooms','beds','price','minimum_nights','maximum_nights','number_of_reviews']
do_listings=pd.read_csv('/Users/chloe/listings.csv')
do_listings=do_listings[features]
do_listings['price']=do_listings.price.str.replace("\$|,","").astype('float') 
# delete the symbol($|,) from price and transfer to float type
do_listings=do_listings.dropna() 
# delete the null elements
do_listings[features]=StandardScaler().fit_transform(do_listings[features])
# use sklearn StandardScaler().fit to make the value in the interval of [-1,1]
normalized_listings=do_listings
print(do_listings.shape)
print(normalized_listings.head())

```

```python
normal_train_df=do_listings.copy().iloc[:2792]# use 75% as traing set
normal_test_df=do_listings.copy().iloc[2792:]
```

```python
from scipy.spatial import distance
'''
parameters:
    new_listing_value: testing set to predict price
    feature_columns: features for futher knn
return:
    predicted price
'''
def predict_price_mul(new_listing_value,feature_columns):
    temp_df=normal_train_df
    temp_df['distance']=distance.cdist(temp_df[feature_columns],[new_listing_value[feature_columns]])# euclidean distance using matrix
    temp_df=temp_df.sort_values('distance')
    knn_5=temp_df.price.iloc[:5] # sort and find the nearest 5 values
    predicted_price=knn_5.mean()
    return predicted_price
'''
perameters:
    predicted_price: from knn method
    price: real value
return:
    mean square error
'''
def prmse(predicted_price,price):
    normal_test_df['sq_error']=(normal_test_df['predicted_price']-normal_test_df['price'])**(2)
    mse=normal_test_df['sq_error'].mean()
    rmse=mse**(1/2)
    return (rmse)
```

```python
cols=['accommodates','bedrooms','beds']
normal_test_df['predicted_price']=normal_test_df[cols].apply(predict_price_mul,feature_columns=cols,axis=1)
rmse=prmse(normal_test_df.predicted_price,normal_test_df.price)
print(rmse)
print(normal_test_df)

```
